```python
# Packages / libraries
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# To install seaborn type "pip install seaborn" to the anaconda terminal

# import sys
# !conda list Check the packages installed
```

```python
raw_data = pd.read_csv('C:\\Users\\TRAN TU VAN\\Desktop\\Python\\Seaborn\\Marketing Raw Data.csv')
raw_data.tail()
```

```python
#Line Graph
#Investigae the trend between date and revenue  

ax = sns.lineplot(x='Week_ID',y='Revenue',data=raw_data)
```

```python
# Example 2 - Adding Categories

# By Promo
ax = sns.lineplot(x='Week_ID',y='Revenue',data=raw_data,hue='Promo')
```

```python
# Example 3 - By Promo with style
#chuyen doi style theo categories
#chi can pass ten cate vao may tu doi
ax = sns.lineplot(x='Week_ID',y='Revenue',data=raw_data,hue='Promo',style='Promo')
```

```python
# Example 4 - By Promo with style & Increase the size & Remove error bars
# Error bar la confidence interval ci = False
# increase the size
sns.set(rc={'figure.figsize':(12,10)})
ax = sns.lineplot(x='Week_ID',y='Revenue',data=raw_data,hue='Promo',style='Promo',ci=False)
```

```python
# Example 5 - By Promo with style & Increase the size & Remove error bars & adding markers & by month
ax = sns.lineplot(x='Month_ID',y='Revenue',data=raw_data,hue='Promo',style='Promo',ci=False,markers=True)
```

```python
# Example 6 - By Day_Name with style & Increase the size & Remove error bars & adding markers & by month
ax = sns.lineplot(x='Month_ID',y='Revenue',data=raw_data,hue='Day_Name',style='Promo',ci=False,markers=True)
```

```python
# Example 7 - By Year with style & Increase the size & Remove error bars & adding markers & by month
#Simplifer == cach doi qua don vi tinh tgian lon hon
#nguoc lai doi sang don vi tinh nho hon
#cua mc s ca x,y
ax = sns.lineplot(x='Year',y='Revenue',data=raw_data,hue='Day_Name',style='Promo',ci=False,markers=True)
```

```python
#Barplot
ax = sns.barplot(x='Month_ID',y='Revenue',data=raw_data)
#Ve lay mean cua cac thang default khong phau sum
#Tu tinh sum voi Index month_ID va value la Revenue
ax1=raw_data[['Revenue','Month_ID']].groupby('Month_ID').agg('sum')
ax1
```

```python
#thanh la confidence interval
#graph duoi chi tinh sum 
sns.barplot(x=ax1.index,y='Revenue',data=ax1)
```

```python
# Example 3 - Mean Revenue by Month - Remove the Confidence Interval - By Promo
ax = sns.barplot(x='Month_ID',y='Revenue',data=raw_data,ci=False,hue='Promo')
```

```python
# Example 4 - Mean Revenue by Month - Remove the Confidence Interval - By Promo - Changing direction
ax = sns.barplot(x='Revenue',y='Year',data=raw_data,ci=False,hue='Promo',orient='h')
```

```python
# Example 5 - Mean Revenue by Month - Remove the Confidence Interval - By Promo - Changing direction - Changing Colour
ax = sns.barplot(x='Revenue',y='Year',data=raw_data,ci=False,hue='Promo',orient='h',color='purple')
```

```python
# Example 1 - Investigating the distribution of Revenue
x=raw_data['Revenue']
sns.distplot(x,color='b')
```

```python
# Example 2 - Investigating the distribution of Revenue, adding the mean
mean=raw_data['Revenue'].mean()
sns.distplot(x,color='b')
#ploting the mean xai matplotlib
plt.axvline(mean,0,1,color='r')

```

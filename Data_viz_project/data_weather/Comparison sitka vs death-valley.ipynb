```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Get dates, high, low temperatures from file
filename = r'C:\Users\TRAN TU VAN\Desktop\Python\Data_viz_project\data_weather\sitka_weather_2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    dates,highs,lows = [],[],[]
    for row in reader:
        try:
            current_date = datetime.strptime(row[0],"%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])
        except ValueError:
            print(current_date,'missing value')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)
            
#Plot data
fig = plt.figure(dpi = 128, figsize=(10,6))
plt.plot(dates,highs,c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
plt.fill_between(dates,highs,lows, facecolor='blue',alpha=0.1)

#format plot.
title = 'Daily high and low temperatures - 2014\nSitka, AK'
plt.title(title,fontsize=20)
plt.xlable('',fontsize=16)
fig.autofmt_xdate()
plt.ylabel('Temperature F',fontsize=16)
plt.tick_params(axis='both',which='major',labelsize=16)
plt.ylim(10,120)

plt.show()
```
